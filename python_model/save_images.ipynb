```python
import torch
from torchvision import datasets, transforms
from PIL import Image
import os
```

```python
# 1. 准备目录
save_dir = "test_images"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 2. 下载/加载 MNIST 测试集
# 注意：这里我们不进行 Normalize，因为我们要保存原始图片(.png)
# Java 端读取图片后，再手动进行 Normalize 处理，这样才符合真实的推理流程
dataset = datasets.MNIST('./data', train=False, download=True)

print(f"开始保存图片到 {save_dir} ...")

# 3. 保存前 10 张图片
for i in range(10):
    img, label = dataset[i] # img 是 PIL Image 对象
    
    # 文件名格式: index_label.png (例如 0_7.png 表示第0张图，数字是7)
    filename = f"{i}_{label}.png"
    save_path = os.path.join(save_dir, filename)
    
    # 调整大小：MNIST 原始是 28x28，但为了方便看，我们可以存大一点，
    # 不过为了配合你的模型，我们最好还是存 28x28，Java 读取时更直接
    img = img.resize((28, 28))
    img.save(save_path)
    print(f"已保存: {filename}")

print("图片生成完毕！")
```
